### (1) ë°ì´í„° ì „ì²˜ë¦¬

```python
# ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜µì‹œë‹¤.
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/mobile_cust_churn.csv'
data = pd.read_csv(path)
```

```python
data.head()
```

* ë³€ìˆ˜ì •ë¦¬ : ë¶ˆí•„ìš”í•œ ë³€ìˆ˜ë¥¼ ì •ë¦¬í•©ì‹œë‹¤.
    * ì‹ë³„ì : ì¼ë ¨ë²ˆí˜¸, ì£¼ë¯¼ë²ˆí˜¸, ì „í™”ë²ˆí˜¸, ê³ ê°ID, ì‚¬ë²ˆ 
    * ì‹œê³„ì—´ ë°ì´í„° ì¤‘ : ì–´ë–¤ ê¸°ê°„ë™ì•ˆ ê±°ì˜ ë³€í™”ê°€ ì—†ëŠ” ê°’.(ë³€ë™ì´ ê±°ì˜ ì—†ëŠ”) ê°’.


```python
drop_cols = ['id']
data.drop(drop_cols, axis = 1, inplace = True )
```

* x, y ë¶„í• 

```python
target = 'CHURN'
x = data.drop(target, axis = 1)
y = data.loc[:, target]
```

* ê°€ë³€ìˆ˜í™” 

```python
dumm_cols = ['REPORTED_SATISFACTION','REPORTED_USAGE_LEVEL','CONSIDERING_CHANGE_OF_PLAN']
x = pd.get_dummies(x, columns = dumm_cols, drop_first = True)
```

* train, val ë¶„í• 

```python
x_train, x_val, y_train, y_val = train_test_split(x,y,test_size = 0.3)
```

### (2) ëª¨ë¸ë§ : ë¡œì§€ìŠ¤í‹± íšŒê·€
* ë¡œì§€ìŠ¤í‹± íšŒê·€ ëª¨ë¸ì„ ë§Œë“¤ê³ 
* ì„±ëŠ¥ í‰ê°€ë¥¼ í•´ ë´…ì‹œë‹¤.

```python
model1 = LogisticRegression()
model1.fit(x_train, y_train)
pred1 = model1.predict(x_val)
```

```python
print(confusion_matrix(y_val, pred1))
print('ğŸ’˜'*20)
print(classification_report(y_val, pred1))
```

### (3) ëª¨ë¸ë§ : KNN
* ìŠ¤ì¼€ì¼ë§ì„ ìˆ˜í–‰
* KNeighborsClassifier ë¥¼ ë¶ˆëŸ¬ì™€ì„œ ëª¨ë¸ë§
* ì„±ëŠ¥ í‰ê°€ë¥¼ í•´ ë´…ì‹œë‹¤.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_train_s = scaler.fit_transform(x_train)
x_val_s = scaler.transform(x_val)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model2 = KNeighborsClassifier(n_neighbors = 50)
model2.fit(x_train_s, y_train)
pred2 = model2.predict(x_val_s)
```

```python
print(confusion_matrix(y_val, pred2))
print('ğŸ’˜'*20)
print(classification_report(y_val, pred2))
```
